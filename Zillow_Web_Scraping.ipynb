```python
import pandas as pd
import json

def extract_data(json_list):
    processed_data = []

    for data in json_list:
        flat_data = {}

        # Top-level fields
        flat_data['zpid'] = data.get('zpid')
        flat_data['status'] = data.get('statusText')
        flat_data['price'] = data.get('price')
        flat_data['beds'] = data.get('beds')
        flat_data['baths'] = data.get('baths')
        flat_data['area'] = data.get('area')
        flat_data['address'] = data.get('address')
        flat_data['brokerName'] = data.get('brokerName')
        flat_data['listingType'] = data.get('listingType')
        flat_data['latitude'] = data.get('latLong', {}).get('latitude')
        flat_data['longitude'] = data.get('latLong', {}).get('longitude')

        # Nested hdpData -> homeInfo
        home_info = data.get('hdpData', {}).get('homeInfo', {})
        flat_data['zipcode'] = home_info.get('zipcode')
        flat_data['city'] = home_info.get('city')
        flat_data['state'] = home_info.get('state')
        flat_data['priceForHDP'] = home_info.get('priceForHDP')
        flat_data['priceChange'] = home_info.get('priceChange')
        flat_data['homeType'] = home_info.get('homeType')
        flat_data['daysOnZillow'] = home_info.get('daysOnZillow')
        flat_data['taxAssessedValue'] = home_info.get('taxAssessedValue')
        flat_data['lotAreaValue'] = home_info.get('lotAreaValue')
        flat_data['lotAreaUnit'] = home_info.get('lotAreaUnit')

        processed_data.append(flat_data)

    # Create DataFrame
    df = pd.DataFrame(processed_data)

    # Handle null values
    df.fillna(value=pd.NA, inplace=True)

    return df

# Example usage
json_list = response.json()['cat1']['searchResults']['listResults']

df = extract_data(json_list)
df
```
